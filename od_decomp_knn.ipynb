```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import plotly.graph_objects as go
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,roc_curve,auc
#from pyod.models.mad import MAD
```

```python
#df=pd.read_csv('pageblocks.csv')
#df=pd.read_csv('satellite-unsupervised-ad.csv')
#df=pd.read_csv('PenDigits.csv')
#df=pd.read_csv('OpticalDigits.csv')
#df=pd.read_csv('SatImage.csv')
#df=pd.read_csv('Waveform.csv')
df=pd.read_csv('ANN-Thyroid.csv')
#df=pd.read_csv('letter-unsupervised-ad.csv')
#df=pd.read_csv('ALOI.csv')
df.shape
```

```python
y=df.iloc[:,-1]
unique, counts = np.unique(y, return_counts=True)
dict_X = dict(zip(unique, counts))
dict_X
```

```python
from sklearn import preprocessing
for x in df.columns:
    while df[x].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(df[x].values))
        df[x] = lbl.transform(list(df[x].values))
```

```python
df.columns
```

```python
result=df.iloc[:,:-1]
data=result.copy()
y_true=df[['outlier']]
```

```python
# analysis of knn
from pyod.models.knn import KNN
knn = KNN()
knn.fit(data)
```

```python
result['Outlier_knn']=knn.predict(data)
y_pred_knn=result[['Outlier_knn']]
y_score_knn=knn.predict_proba(data)[:,1]
confusion_matrix(y_true,y_pred_knn)
```

```python
#KNN results without decomposition
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_true, y_score_knn,pos_label=1)
print('Precision: %.3f' % precision_score(y_true, y_pred_knn))
print('Recall: %.3f' % recall_score(y_true, y_pred_knn))
print('f1_score: %.3f' % f1_score(y_true, y_pred_knn))
print('AUC: %.3f' % auc(fpr_knn, tpr_knn))
```

```python
cluster = KMeans(n_clusters=5)
cluster.fit_predict(data)
```

```python
df['cluster']=pd.Categorical(cluster.labels_)

cluster_1=df.query('cluster==0')
cluster_2=df.query('cluster==1')
cluster_3=df.query('cluster==2')
cluster_4=df.query('cluster==3')
cluster_5=df.query('cluster==4')
```

```python
dt_1=cluster_1.iloc[:,:-1]
dt_2=cluster_2.iloc[:,:-1]
dt_3=cluster_3.iloc[:,:-1]
dt_4=cluster_4.iloc[:,:-1]
dt_5=cluster_5.iloc[:,:-1]
```

```python
#iteration_1
data_1=cluster_1.iloc[:,:-2]
data_2=cluster_2.iloc[:,:-2]
data_3=cluster_3.iloc[:,:-2]
data_4=cluster_4.iloc[:,:-2]
data_5=cluster_5.iloc[:,:-2]

data_1.reset_index(drop=True, inplace=True)
data_2.reset_index(drop=True, inplace=True)
data_3.reset_index(drop=True, inplace=True)
data_4.reset_index(drop=True, inplace=True)
data_5.reset_index(drop=True, inplace=True)

print('data points in cluster_1 initeration_1::', data_1.shape[0])
print('data points in cluster_2 initeration_1::', data_2.shape[0])
print('data points in cluster_3 initeration_1::', data_3.shape[0])
print('data points in cluster_4 initeration_1::', data_4.shape[0])
print('data points in cluster_5 initeration_1::', data_5.shape[0])
```

```python
#cluster_1 outlier detection
knn1 = KNN()
knn1.fit(data_1)
```

```python
#cluster_2 outlier detection
knn2 = KNN()
knn2.fit(data_2)
```

```python
#cluster_3 outlier detection
knn3 = KNN()
knn3.fit(data_3)
```

```python
#cluster_4 outlier detection
knn4 = KNN()
knn4.fit(data_4)
```

```python
#cluster_4 outlier detection
knn5 = KNN()
knn5.fit(data_5)
```

```python
#od results
dt_1['od']=knn1.predict(data_1)
dt_2['od']=knn2.predict(data_2)
dt_3['od']=knn3.predict(data_3)
dt_4['od']=knn4.predict(data_4)
dt_5['od']=knn5.predict(data_5)

dt_1['prob']=knn1.predict_proba(data_1)[:,1]
dt_2['prob']=knn2.predict_proba(data_2)[:,1]
dt_3['prob']=knn3.predict_proba(data_3)[:,1]
dt_4['prob']=knn4.predict_proba(data_4)[:,1]
dt_5['prob']=knn5.predict_proba(data_5)[:,1]
```

```python
#combine the cluster
com=pd.concat([dt_1,dt_2,dt_3,dt_4,dt_5],axis=0)
com.reset_index(drop=True, inplace=True)
com.shape
```

```python
y_tr=com[['outlier']]
y_pr=com[['od']]
y_sc=com[['prob']]
confusion_matrix(y_tr,y_pr)
```

```python
#KNN results with decomposition(k=2)
fpr_com1, tpr_com1, thresholds_com1 = roc_curve(y_tr, y_sc,pos_label=1)
print('Precision: %.3f' % precision_score(y_tr, y_pr))
print('Recall: %.3f' % recall_score(y_tr, y_pr))
print('f1_score: %.3f' % f1_score(y_tr, y_pr))
print('AUC: %.3f' % auc(fpr_com1, tpr_com1))
```

```python
#KNN results with decomposition(k=3)
fpr_com2, tpr_com2, thresholds_com2 = roc_curve(y_tr, y_sc,pos_label=1)
print('Precision: %.3f' % precision_score(y_tr, y_pr))
print('Recall: %.3f' % recall_score(y_tr, y_pr))
print('f1_score: %.3f' % f1_score(y_tr, y_pr))
print('AUC: %.3f' % auc(fpr_com2, tpr_com2))
```

```python
#KNN results with decomposition(k=4)
fpr_com3, tpr_com3, thresholds_com3 = roc_curve(y_tr, y_sc,pos_label=1)
print('Precision: %.3f' % precision_score(y_tr, y_pr))
print('Recall: %.3f' % recall_score(y_tr, y_pr))
print('f1_score: %.3f' % f1_score(y_tr, y_pr))
print('AUC: %.3f' % auc(fpr_com3, tpr_com3))
```

```python
#KNN results with decomposition(k=5)
fpr_com4, tpr_com4, thresholds_com4 = roc_curve(y_tr, y_sc,pos_label=1)
print('Precision: %.3f' % precision_score(y_tr, y_pr))
print('Recall: %.3f' % recall_score(y_tr, y_pr))
print('f1_score: %.3f' % f1_score(y_tr, y_pr))
print('AUC: %.3f' % auc(fpr_com4, tpr_com4))
```

```python
#ROC analysis
import plotly.graph_objects as go

fig = go.Figure()
fig.add_shape(
    type='line', line=dict(dash='dash'),
    x0=0, x1=1, y0=0, y1=1)

fig.add_trace(go.Scatter(x=fpr_knn, y=tpr_knn, name=f'kNN(AUC={auc(fpr_knn, tpr_knn):.4f})', mode='lines',marker_color='crimson'))
fig.add_trace(go.Scatter(x=fpr_com1, y=tpr_com1, name=f'AD(k=2)(AUC={auc(fpr_com1, tpr_com1):.4f})', mode='lines',marker_color='hotpink'))
fig.add_trace(go.Scatter(x=fpr_com2, y=tpr_com2, name=f'AD(k=3)(AUC={auc(fpr_com2, tpr_com2):.4f})', mode='lines'))
fig.add_trace(go.Scatter(x=fpr_com3, y=tpr_com3, name=f'AD(k=4)(AUC={auc(fpr_com3, tpr_com3):.4f})', mode='lines'))
fig.add_trace(go.Scatter(x=fpr_com4, y=tpr_com4, name=f'AD(k=5)(AUC={auc(fpr_com4, tpr_com4):.4f})', mode='lines'))

fig.update_layout(
    xaxis_title='False Positive Rate',
    yaxis_title='True Positive Rate',
    yaxis=dict(scaleanchor="x", scaleratio=1),
    xaxis=dict(constrain='domain'),
    width=700, height=600)
fig.show()
```
